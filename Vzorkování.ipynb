```python
import matplotlib.patches as mpp
from scipy.fftpack import fft, ifft
##########################################3

#  Konstanty
fs=8e3
Ts=1./fs
length=1

fmin=5.
Tmin=1./fmin
# časová osa
t = arange(0,length,1./fs)
u = 1*cos(2*pi*5*t)+1.2*cos(2*pi*6*t)+0.9*cos(2*pi*7*t)+0.6*cos(2*pi*8*t)+0.4*cos(2*pi*9*t)

tvz=t[::200]
uvz=u[::200]

## Spektrum
Ck= fft(u)
absCk = 2./len(u) * numpy.abs(Ck)[:len(Ck)/2]
f = arange(0,fs/2,float(fs)/len(Ck))

Svz= fft(uvz)
absSvz = 2./len(uvz) * numpy.abs(Svz)[:len(Svz)/2]
fvz = arange(0,fs/2/200,float(fs/200)/len(Svz))

# Modulační signá
fig=figure()
subplot(211)
plot(t,u)
stem(tvz, uvz,'b-','ro','x')
title(u'Modulační signál')
grid('on')
xlabel(r't\,[s] $\rightarrow$', x=0.9)
ylabel('$\mathrm u_m$\,[V] $\uparrow$', y=0.9, rotation=0)
ylim([-1.1*max(u), 1.1*max(u)])
xlim([0,3*Tmin])

subplot(212)
title(ur'Modulační signál -- amplitudové spektrum')
plot(f,absCk,'.-')
plot(fvz,absSvz,'r.-')
xlabel(r'f\,[Hz] $\rightarrow$', x=0.9)
ylabel('$\mathrm U_m$\,[V] $\uparrow$', y=0.9, rotation=0)
ylim([0,1.1*max(absCk)]);
xlim([0,50])
grid(1)
#text(500,-0.2 , '$\mathbf{f_m}$', size=16)
#figtext(0,1 , '$\mathrm f_m$', size=16)
tight_layout()
```
