Amplitudová modulace
======================

 * nastavím konstanty

```python
fs=8e3
Ts=1./fs
length=2

Umax=1
f1=float(100)
T1=1./f1
```

```python
t = arange(0,length,1./fs)
u = Umax*cos(2*pi*f1*t)
```

* kreslím časovou oblas u

```python
plot(t,u)
title('Modulacni signal')
grid('on')
xlabel('t[s]')
ylabel('u[V]')
ylim([-1.1*Umax, 1.1*Umax])
xlim([0,3*T1])
```

* kreslím frekvenční oblast u

```python
import pylab as lab
Ck= lab.fft(u)
absCk = 2*abs(Ck)[:len(Ck)/2]
f = lab.arange(0,fs/2,float(fs)/len(Ck))
```

```python
plot(f,absCk,'-+')
ylim([0,1.1*max(absCk)]);
```
