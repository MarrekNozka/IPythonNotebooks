Amplitudová modulace
======================

```python
#  Konstanty
fs=44100
Ts=1./fs
length=1
# modulační vlna
fm=500.
Tm=1./fm
# nosná vlna
m=0.8
fc=6000.
Tc=1./fs
Uc=3
# časová osa
t = arange(0,length,1./fs)

# modulační signál
um = 1*cos(2*pi*fm*t)

# převod do frekvenční oblasti
from scipy.fftpack import fft, ifft
Ck= fft(um)
absCk = 2./len(um) * numpy.abs(Ck)[:len(Ck)/2]
f = arange(0,fs/2,float(fs)/len(Ck))
# Modulační signá
figure()
subplot(211)
plot(t,um)
title(u'Modulační signál')
grid('on')
xlabel(r't\,[s] $\rightarrow$', x=0.9)
ylabel('u\,[V] $\uparrow$', y=0.9, rotation=0)
ylim([-1.1*max(um), 1.1*max(um)])
xlim([0,3*Tm])

subplot(212)
plot(f,absCk,'.-')
xlabel(r'f\,[Hz] $\rightarrow$', x=0.9)
ylabel('U\,[V] $\uparrow$', y=0.9, rotation=0)
ylim([0,1.1*max(absCk)]);
xlim([0,10e3])
grid(1)
```

```python
# Nosná vlna před modulací

uc = Uc*cos(2*pi*fc*t)

Ck= fft(uc)
absCk = 2./len(uc) * abs(Ck)[:len(Ck)/2]
f = arange(0,fs/2,float(fs)/len(Ck))

figure()
subplot(211)
plot(t,uc)
title(u'Nosná vlana před modulací')
grid('on')
xlabel(r't\,[s] $\rightarrow$', x=0.9)
ylabel('u\,[V] $\uparrow$', y=0.9, rotation=0)
ylim([-1.1*max(uc), 1.1*max(uc)])
xlim([0,3*Tm])

subplot(212)
plot(f,absCk,'.-')
xlabel(r'f\,[Hz] $\rightarrow$', x=0.9)
ylabel('U\,[V] $\uparrow$', y=0.9, rotation=0)
ylim([0,1.1*max(absCk)]);
xlim([0,10e3])
grid(1)
```

```python
# Nosná vlna po modulací
m=0.8
uc = Uc*(1+m*um)*cos(2*pi*fc*t)

Ck= fft(uc)
absCk = 2./len(uc) * abs(Ck)[:len(Ck)/2]
f = arange(0,fs/2,float(fs)/len(Ck))

figure()
subplot(211)
plot(t,uc)
title(ur'Nosná vlana po modulaci m=80\%')
grid('on')
xlabel(r't\,[s] $\rightarrow$', x=0.9)
ylabel('u\,[V] $\uparrow$', y=0.9, rotation=0)
ylim([-1.1*max(uc), 1.1*max(uc)])
xlim([0,3*Tm])

subplot(212)
plot(f,absCk,'.-')
xlabel(r'f\,[Hz] $\rightarrow$', x=0.9)
ylabel('U\,[V] $\uparrow$', y=0.9, rotation=0)
ylim([0,1.1*max(absCk)]);
xlim([0,10e3])
grid(1)
```
